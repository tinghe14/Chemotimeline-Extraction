```python
import json 
import os 
import pandas as pd 
import numpy as np 
import random
from collections import defaultdict
```

```python
def _get_size(cancer, mode):
    if (cancer == "breast" or cancer == "ovarian") and mode == "train":
        return 100
    elif cancer == "melanoma" and mode == "train":
        return 10 
    elif (cancer == "breast" or cancer == "ovarian") and mode == "dev":
        return 20 
    elif cancer == "melanoma" and mode == "dev":
        return 2 
    
```

```python
# sample

# exclude None in the JSON
# random choose, given len generate x random number
    ## train: o/b/m: 100, 100, 10
    ## dev:20, 2, 20
    ## 1/5, 2GPU, 5CPU, running time: 
# select and change to the require format
mode = "dev"


final_lst = []
for cancer in ["breast", "ovarian", "melanoma"]:
    file = f"/users/the/NER_MTB/timeset/my_code_result/input_data/tf_rel_task/temporal-nli/{cancer}_{mode}_pos_neg_for_nli.json"
    with open(file, "r") as infile: 
        dct = json.load(infile)

    non_empty_rel = []
    for pair in dct: 
        if (pair["arg1"]["start"] is None) or (pair["arg1"]["end"] is None) or (pair["arg2"]["start"] is None) or (pair["arg1"]["end"] is None): 
            continue 
        non_empty_rel.append(pair)

    with open(f'./{cancer}_{mode}_non_empty_rel.json', 'w') as f:
        json.dump(non_empty_rel, f, indent=4) 
    len(non_empty_rel)

    size = _get_size(cancer, mode)
    print(size)
    random_lst = np.random.randint(0, len(non_empty_rel), size)
    print(len(non_empty_rel), random_lst)

    for ind, pair in enumerate(non_empty_rel):
        if ind in random_lst:
            chose_pair = defaultdict()
            chose_pair["filename"] = pair["filename"]
            chose_pair["context"] = pair["context"]
            chose_pair["statement"] = pair["statement"]
            chose_pair["arg1"] = {"mention": pair["arg1"]["mention"],
                                  "start": pair["arg1"]["start"],
                                   "end": pair["arg1"]["end"]}
            chose_pair["arg2"] = {"mention": pair["arg2"]["mention"],
                            "start": pair["arg2"]["start"],
                            "end": pair["arg2"]["end"]}
            chose_pair["label"] = pair["label"]
            final_lst.append(chose_pair)
        else: 
            continue

with open(f'/users/the/NER_MTB/timeset/my_code_result/input_data/tf_rel_task/temporal-nli/{mode}.json', 'w') as f:
        json.dump(final_lst, f, indent=4) 
```
